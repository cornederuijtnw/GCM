```python
import numpy as np
import pandas as pd
import tensorflow as tf
import tensorflow_probability as tfp
from tensorflow.keras.initializers import Zeros
from IPython.display import display
```

```python
np.random.seed(1992)

N = 11
a = np.zeros((N, N))
w = np.random.normal(0, 1, int(N*(N-1)/2)) # -1 as we remove the diagonal
upper_indices = np.vstack(np.triu_indices(N, k=1))

upper_indices = upper_indices[0, :] * N + upper_indices[1, :]
np.put(a, upper_indices, w)

display(pd.DataFrame(a))
```

```python
N = 11

w_init = tf.random_normal_initializer()
zero_init = tf.zeros_initializer()

w = tf.Variable(initial_value = w_init(shape=(int(N*(N-1)/2), ), dtype='float32'), trainable=True)
zero_col_var = tf.Variable(initial_value = zero_init(shape=(N, 1), dtype='float32'), trainable=False) 
zero_row_var = tf.Variable(initial_value = zero_init(shape=(1, N-1), dtype='float32'), trainable=False)

w = tfp.math.fill_triangular(w, upper=True)
w = tf.concat([w, zero_row_var], axis=0)
w = tf.concat([zero_col_var, w], axis=1)

inputs = tf.constant(np.eye(N), dtype='float32')
display(pd.DataFrame(inputs.numpy()))

inputs = tf.tile(inputs, [N, 1])

display(pd.DataFrame(w.numpy()))
display(pd.DataFrame(inputs.numpy()))

# logis = tf.multiply(inputs, w)


# display(pd.DataFrame(np.triu(inputs, k=1)))

# input_mask = tf.linalg.band_part(inputs, )

# res = tf.transpose(tf.transpose(tf.exp(logis))/tf.reduce_sum(tf.exp(logis), axis=1))

# display(pd.DataFrame(res.numpy()))




















```
